```r
# Here, I am loading the necessary libraries
library(ggplot2)
```

```r
# Here, I am loading the data
Netflix_shows_movies <- read.csv("netflix_data.csv")
```

```r
# Here, I am plotting the top 5 Most Watched Genres
top_genres <- head(sort(table(Netflix_shows_movies$listed_in), decreasing = TRUE), 5)
top_genres_df <- data.frame(Genre = names(top_genres), Count = as.numeric(top_genres))

# I am reordering the Genre based on Count
top_genres_df$Genre <- factor(top_genres_df$Genre, levels = top_genres_df$Genre[order(top_genres_df$Count, decreasing = TRUE)])

ggplot(top_genres_df, aes(x = Genre, y = Count)) +
    geom_bar(stat = "identity", fill = "steelblue") +
    theme_minimal() +
    labs(title = "Top 5 Genres with the Most Shows / Movies", x = "Genre", y = "Count") +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
